```python
''' class designed to create a new chart featuring the top 25 radioless songs of the week, in the US'''

def Uninstantiated(hot100, radiosongs):
    current_rank = 1
    UninstantiatedChart = []
    for i in range(len(hot100)):
        current_song = hot100[i].title
        current_artist = hot100[i].artist
        uninstantiatedfound = 1
        for j in range(len(radiosongs) - 10):
            temp_song = radiosongs[j].title
            temp_artist = radiosongs[j].artist
            if (temp_song == current_song) and ((temp_artist in current_artist) or (current_artist in temp_artist)):
                uninstantiatedfound -= 1
                break
        
        if uninstantiatedfound == 1:
            UninstantiatedChart.append(current_song)
            current_rank += 1
        else:
            continue
        
        if current_rank == 26:
            break
    return UninstantiatedChart


def Repr(chart):
    for i in range(len(chart)):
        print(str(i + 1) + " " + chart[i], end='\n')
```

```python
Zeroplay = Uninstantiated(chart, chart2)
Repr(Zeroplay)
```

```python
chart3 = billboard.ChartData('billboard-200')
album = chart3[55]

def create_album_roadtrip(billboard200):
    RoadTripChart = []
    for i in range(len(billboard200)):
        rank = i + 1
        current_album = billboard200[i].title
        album_weeks = billboard200[i].weeks
        if (billboard200[i].lastPos == 0) and album_weeks > 1:
            rank *= 1.5
        if float(album_weeks) < 4.0:
            rankmultiplier = 4.0 / album_weeks
            rank *= rankmultiplier
        if float(album_weeks) > 100.0:
            agepenalty = (album_weeks - 100.0) / 4.0
            rank += agepenalty
        RoadTripChart.append((current_album, float(rank)))
    RoadTripChart.sort(key=lambda tup: tup[1])
    return RoadTripChart
        

result = create_album_roadtrip(chart3)
for i in range(len(result)):
    print(str(i + 1) + ":", result[i])
```
